El objetivo es entrenar una red neuronal para que aprenda esta conversiГіn.

##GeneraciГіn de datos de entrenamiento

```python
celsius = np.array([-35,-30, -20,-10, -5, 0, 10, 15, 20, 25, 35,40], dtype=float)
fahrenheit = celsius * 1.8 + 32
fahrenheit
```

#CreaciГіn del modelo de red neuronal

* Para aplicar redes neuronales, vamos a utilizar la librerГӯa Keras, que nos facilita el desarrollo de modelos de redes neuronales de manera sencilla. En este caso, vamos a crear una variable llamada layer de tipo Layer en Keras. Las capas densas son aquellas en las que cada neurona estГЎ conectada con todas las neuronas de la siguiente capa. En este modelo, tenemos dos capas: un tensor de entrada y otra con una neurona de salida. Al definir la capa, pasamos como parГЎmetro units, que representa el nГәmero de neuronas en la capa de salida (en este caso, 1). AdemГЎs, usamos el parГЎmetro **input_shape** con valor 1, ya que tenemos un solo Tensor (celsius) de entrada.

* A continuaciГіn, creamos una variable de objeto llamada model de tipo Sequential en Keras para agrupar y organizar las capas. El modelo Sequential es el mГЎs sencillo y se utiliza para redes neuronales en las que las capas se apilan una sobre otra de forma lineal. Para redes mГЎs complejas, existen otros tipos de modelos.

* DespuГ©s de definir el modelo, es necesario compilarlo para que pueda ser entrenado. Para ello, se pasan dos parГЎmetros:

* **optimizer**: AquГӯ usamos el algoritmo Adam, que ayuda a que la red neuronal aprenda de manera eficiente, ajustando los pesos y sesgos para mejorar la predicciГіn con cada iteraciГіn. La tasa de aprendizaje (learning rate) es un valor clave que determina quГ© tan grande es el ajuste de los pesos y sesgos en cada paso. En general, la tasa de aprendizaje se elige mediante prueba y error, pero un buen punto de partida es un valor pequeГұo, como 0.1 o 0.01. Si el valor es muy pequeГұo, el ajuste serГЎ muy lento; si es muy grande, la red puede no converger correctamente y no aprender bien.

* **loss function**: Usamos la funciГіn de pГ©rdida mean_squared_error (error cuadrГЎtico medio). Esta funciГіn penaliza los errores grandes de manera mГЎs severa que los pequeГұos, lo que significa que una pequeГұa cantidad de errores grandes es peor que muchos errores pequeГұos.


```python
# Fijar la semilla pseudoaleatoria para TensorFlow
tf.random.set_seed(42)
```

```python

# Creamos el modelo de red neuronal con Input() como la capa de entrada
optmizador = tf.keras.optimizers.Adam(learning_rate=0.1)
model = tf.keras.Sequential([
    Input(shape=(1,)),  # Define la forma de entrada como un solo valor
    tf.keras.layers.Dense(1)  # Capa densa con 1 neurona de salida
])

# Compilamos el modelo con el optimizador 'adam' y la funciГіn de pГ©rdida 'mean_squared_error'
model.compile(optimizer=optmizador, loss='mean_squared_error')

#Pruebe modificando el optimzador, solo poner optimizer='adam'
```

```python
model.summary()
```

#Entrenamiento del modelo
Luego creamos el historial, le pasamos la lista de entrada y de salida, las vueltas (iteraciones) y verbose=false para que no imprima cosas que no necesitamos

```python
print("Comienza entrenamiento..")
history = model.fit(celsius, fahrenheit, epochs=1000, verbose=True)
print("Modelo entrenado!")

#Pruebe poniendo verbose=False, menos epochs.
```

#Visualizar la funciГіn de pГ©rdida

DespuГ©s, usamos matplotlib para poder graficar el comportamiento de aprendizaje de la red en base a las vueltas y las capas. Podemos observar que la red necesitГі unas 400 vueltas aprox para encontrar el algoritmo Гіptimo para el problema.

```python
# Visualizamos la evoluciГіn de la pГ©rdida durante el entrenamiento
plt.plot(history.history['loss'])
plt.title('EvoluciГіn de la pГ©rdida durante el entrenamiento')
plt.xlabel('Гүpocas')
plt.ylabel('PГ©rdida')
plt.show()
```

```python
print("PrecciГіn 1")
result = model.predict(np.array([-27,-4,8,16])).round(1)
print("El resultado es" + str(result) + " fahrenheit.")
```

```python
#Mostar los pesos y el bias
print(model.get_weights())
```

#Evaluar grГЎficamente el modelo con los datos reales

```python
# Evaluamos el modelo con los mismos datos de entrada para ver el rendimiento
predicciones = model.predict(celsius)

# Visualizamos las predicciones junto con los datos reales
plt.scatter(celsius, fahrenheit, color='blue', label='Datos reales')
plt.scatter(celsius, predicciones, color='red', label='Predicciones del modelo')
plt.title('Predicciones vs Datos reales')
plt.xlabel('Grados Celsius')
plt.ylabel('Grados Fahrenheit')
plt.legend()
plt.show()
```
