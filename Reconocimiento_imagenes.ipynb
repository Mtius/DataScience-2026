```python
# Sobel horizontal
sobel_h = np.array([[-1, -2, -1],
                    [ 0,  0,  0],
                    [ 1,  2,  1]])

# Sobel vertical
sobel_v = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

# Gaussiano
gauss = (1/16) * np.array([[1, 2, 1],
                           [2, 4, 2],
                           [1, 2, 1]])

# Repujado (Emboss)
emboss = np.array([[-2, -1, 0],
                   [-1,  1, 1],
                   [ 0,  1, 2]])
```

```python
# Usar imagen en formato OpenCV
fortnite_cv = cv2.cvtColor(fortnite_img, cv2.COLOR_RGB2BGR)

img_original = fortnite_img

img_sobel_h = cv2.filter2D(fortnite_cv, -1, sobel_h)
img_sobel_v = cv2.filter2D(fortnite_cv, -1, sobel_v)
img_gauss = cv2.filter2D(fortnite_cv, -1, gauss)
img_emboss = cv2.filter2D(fortnite_cv, -1, emboss)

# Convertir a RGB
img_sobel_h = cv2.cvtColor(img_sobel_h, cv2.COLOR_BGR2RGB)
img_sobel_v = cv2.cvtColor(img_sobel_v, cv2.COLOR_BGR2RGB)
img_gauss = cv2.cvtColor(img_gauss, cv2.COLOR_BGR2RGB)
img_emboss = cv2.cvtColor(img_emboss, cv2.COLOR_BGR2RGB)

# Mostrar resultados
plt.figure(figsize=(12,8))

plt.subplot(2,3,1)
plt.imshow(img_original)
plt.title("Original")
plt.axis('off')

plt.subplot(2,3,2)
plt.imshow(img_sobel_h)
plt.title("Sobel Horizontal")
plt.axis('off')

plt.subplot(2,3,3)
plt.imshow(img_sobel_v)
plt.title("Sobel Vertical")
plt.axis('off')

plt.subplot(2,3,4)
plt.imshow(img_gauss)
plt.title("Gaussiano")
plt.axis('off')

plt.subplot(2,3,5)
plt.imshow(img_emboss)
plt.title("Repujado")
plt.axis('off')

plt.show()
```

```python
# Dimensiones originales
h, w, c = fortnite_img.shape

# Nueva imagen (mitad de tamaño)
pooled_img = np.zeros((h//2, w//2, c), dtype=fortnite_img.dtype)

# Aplicar Max Pooling 2x2
for i in range(0, h, 2):
    for j in range(0, w, 2):
        bloque = fortnite_img[i:i+2, j:j+2]
        pooled_img[i//2, j//2] = bloque.max(axis=(0,1))

# Mostrar resultado
plt.figure(figsize=(5,5))
plt.imshow(pooled_img)
plt.title("Max Pooling 2x2")
plt.axis('on')
plt.show()
```

```python
# Convertir a formato OpenCV (BGR)
fortnite_cv = cv2.cvtColor(fortnite_img, cv2.COLOR_RGB2BGR)

# Kernels Sobel
sobel_h = np.array([[-1, -2, -1],
                    [ 0,  0,  0],
                    [ 1,  2,  1]])

sobel_v = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

# Aplicar Sobel
img_sobel_h = cv2.filter2D(fortnite_cv, -1, sobel_h)
img_sobel_v = cv2.filter2D(fortnite_cv, -1, sobel_v)

# Convertir a RGB
img_sobel_h = cv2.cvtColor(img_sobel_h, cv2.COLOR_BGR2RGB)
img_sobel_v = cv2.cvtColor(img_sobel_v, cv2.COLOR_BGR2RGB)

# -------- MAX POOLING 3x3 --------
def max_pooling_3x3(img):
    h, w, c = img.shape
    pooled = np.zeros((h//3, w//3, c), dtype=img.dtype)

    for i in range(0, h - 2, 3):
        for j in range(0, w - 2, 3):
            bloque = img[i:i+3, j:j+3]
            pooled[i//3, j//3] = bloque.max(axis=(0,1))

    return pooled

# Aplicar pooling
pool_h = max_pooling_3x3(img_sobel_h)
pool_v = max_pooling_3x3(img_sobel_v)

# -------- MOSTRAR RESULTADOS --------
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(img_sobel_h)
plt.title("Sobel Horizontal")
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(img_sobel_v)
plt.title("Sobel Vertical")
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(pool_h)
plt.title("Max Pooling 3x3 (Sobel H)")
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(pool_v)
plt.title("Max Pooling 3x3 (Sobel V)")
plt.axis('off')

plt.show()
```
